<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
states= states.dropna()

states["REVENUE_PER_STUDENT"] = states["TOTAL_REVENUE"] / states["ENROLL"]
states["FED_REVENUE_%"] = (states["FEDERAL_REVENUE"] / states["TOTAL_REVENUE"]) * 100
states["STATE_REVENUE_%"] = (states["STATE_REVENUE"] / states["TOTAL_REVENUE"]) * 100
states["LOCAL_REVENUE_%"] = (states["LOCAL_REVENUE"] / states["TOTAL_REVENUE"]) * 100
states["expenditure_per_student"] = states["TOTAL_EXPENDITURE"] / states["ENROLL"]
states["INSTRUCT_EXPEND_%"] = (states["INSTRUCTION_EXPENDITURE"] / states["TOTAL_EXPENDITURE"]) * 100
states["SUPPORT_EXPEND_%"] = (states["SUPPORT_SERVICES_EXPENDITURE"] / states["TOTAL_EXPENDITURE"]) * 100
states["math_improvement"] = states["AVG_MATH_8_SCORE"] - states["AVG_MATH_4_SCORE"]
states
```

Feature engineering justification: **<BRIEFLY DESCRIBE WHY YOU MADE THE CHANGES THAT YOU DID\>**

```python
#One of the data transformations I made was to calculate the percentages of how each state's revenue was distributed as well as 
#how their expenditure was distributed. A characteristics I thought were important especially was the local revenue percentages 
#for each state because I hypothesized that more local funding could provide a more tailored approach to address specific needs 
#in a community/school district. I also added the expenditure per student by simplying dividing the state's expenditure with the
#number of students to analyze funding quantitatively. Another another statistic created I found potentially useful was creating
#the instruction expenditure percentage, that is, the amount of expenses diverted directly towards teaching/instruction, which I 
#hypothesized would have the most direct impact on test scores. Finally, I also added an improvement feature that tracks the 
#difference in math test score between 4th and 8th grade which might be a better indicator of teaching (what the students 
#actually learned/improved).
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = states[["expenditure_per_student", "math_improvement", "INSTRUCT_EXPEND_%"]]
y = states.loc[X.index]["AVG_MATH_8_SCORE"]

y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python


print(model.intercept_)
print(model.coef_)
```

```python
model.score(X_test, y_test)
#R-squared -> Weak to moderate fit
```

```python
numpy.mean(model.predict(X_test)-y_test)
#Mean error -> moderate fit
```

```python
numpy.mean((model.predict(X_test)-y_test)**2)**0.5
#Root mean squared error -> Heavily affected by outliers
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'math_improvement'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'expenditure_per_student'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_MATH_8_SCORE')
plt.title("Model Behavior on Testing Set")
```
